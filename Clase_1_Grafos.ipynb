Cada grafo, nodo, arista puede almacenar propiedades llave valor. Por default estos valores estan vacios, pero se pueden agregar en diferentes etapas del modelado de los grafos.

# Nodo
Nodos (con atributos) pueden ser agregados de una forma muy sencilla. Esto es de gran utilidad, ya que un grafo puede estar agrgando constantemente nuevos nodos.

```python
# Agregar un nodo con atributos a nivel nodo.
G.add_node(0, feature=0, label=0)

# Obtener atributos de el nodo 0
node_0_attr = G.nodes[0]
print("Atributos del Nodo 0  {}".format(node_0_attr))
```

```python
# Agregar multiples nodos con atributos.
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
])

# Iterar sobre todos los nodos
# Asignar el valor true data regresara atributos del nodo.
for node in G.nodes(data=True):
  print("Atributos del Nodo {}  {}".format(node[0],node[1]))

#Obtener el numero de nodos.
num_nodes = G.number_of_nodes()
print("G contiene {} nodos".format(num_nodes))
```

# Edge
Similar a los nodos, edges(con propiedades) se pueden utilizar para realizar conexiones con otros nodos. Normalmente se agregan estos vinculos con pesos especificando el valor de la relación. Pero esto depende en su mayor medida del modelado del grafo.

```python
# Agrega un vinculo con un peso de 0.5
G.add_edge(0, 1, weight=0.5)

# Obtener los atributos del vinculo (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Edge (0, 1) tiene los atributos {}".format(edge_0_1_attr))
```

```python
# Agrea multiples vinculos con peso.
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Itera todos los vinculos.
# Edges, no tiene la propuedad data=True especificada, por deafult es False, entonces solamente sera regresado el vinculo.

for edge in G.edges():
  print(edge)

# Obtener el numero de vinculos.
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
```

# Visualizando un grafo.
Los grafos son una potente herramienta de análisis, que nos permite intuitivamente concer un poco mas acerca de la red y su complejidad de la misma, aunque mucho de su potencial no radica en la visualización del mismo.

## Ventajas
* Interpretación Intuitiva: Los grafos permiten una representación visual que facilita la comprensión de estructuras complejas y relaciones entre nodos, lo que puede ser más difícil de percibir en formatos de datos tabulares o textuales.
* Detección de Patrones: Facilita la identificación de patrones, como comunidades o módulos, cuellos de botella en redes de comunicación, y nodos influyentes.
* Análisis de Conectividad: Permite analizar cómo los elementos de un sistema están interconectados. Esto es esencial en áreas como la teoría de redes, donde la conectividad puede influir en la robustez y eficiencia de la red.

## Desventajas
* Escalabilidad: A medida que el tamaño del grafo aumenta, su visualización puede volverse confusa y menos informativa. Grafos con miles de nodos y aristas pueden resultar en una visualización desordenada que es difícil de interpretar.
* Dependencia del Diseño: La eficacia de una visualización de grafos depende en gran medida de la elección de su diseño y algoritmo de diseño gráfico. Un mal diseño puede llevar a interpretaciones erróneas o a la omisión de insights importantes.
* Costo Computacional: Los algoritmos para visualizar grafos, especialmente aquellos que intentan optimizar la estética o la disposición, pueden ser computacionalmente costosos, lo cual es un problema con grafos muy grandes.
* Sobrecarga de Información: Si se incluyen demasiados detalles o anotaciones, la visualización puede resultar abrumadora, dificultando la identificación de los aspectos más relevantes del grafo.

# Distribución de grados

La distribución de grados de un grafo es una forma de describir cómo se distribuyen las conexiones entre los nodos en un grafo.

## Grafos No Dirigidos

La **distribución de grados** en este tipo de grafos es un recuento de cuántos nodos tienen cada posible grado. Por ejemplo, si tienes un grafo con 10 nodos y cinco de estos nodos tienen cada uno 3 conexiones, dos nodos tienen 2 conexiones y los otros tres nodos tienen 1 conexión, la distribución de grados te dirá que hay cinco nodos de grado 3, dos nodos de grado 2, y tres nodos de grado 1. Esto te ayuda a entender si la mayoría de los nodos están igualmente conectados o si algunos nodos son mucho más centrales (conectados) que otros.

## Grafos Dirigidos
En un grafo dirigido, las aristas tienen una dirección, indicando una relación de un nodo a otro. Por ejemplo, en una red social, un "seguir" sería una arista dirigida del seguidor al seguido. Aquí, cada nodo tiene dos tipos de grados:
- **Grado de entrada**: Número de aristas que entran al nodo.
- **Grado de salida**: Número de aristas que salen del nodo.

La distribución de grados en grafos dirigidos por lo tanto se puede desglosar en dos:
- **Distribución de grados de entrada**: Muestra cuántos nodos reciben una cantidad específica de conexiones entrantes.
- **Distribución de grados de salida**: Muestra cuántos nodos envían una cantidad específica de conexiones salientes.

Estas distribuciones pueden decirte, por ejemplo, quiénes son los principales influenciadores en una red social (nodos con alto grado de entrada) o quiénes son los principales consumidores de información (nodos con alto grado de salida).

La distribución de grados es una herramienta estadística que describe cuán conectados están los nodos en un grafo, ayudando a identificar patrones como centralidad, dispersión y puntos de influencia en la red, ya sea en un contexto dirigido o no dirigido.

# Densidad de la RED
La densidad de la red es una medida que indica qué tan completa o conectada está una red. En términos técnicos, la densidad proporciona la proporción de conexiones posibles que están efectivamente presentes entre los nodos de la red.

### Cómo se Calcula

Para redes no dirigidas, la densidad se calcula usando la fórmula:
$$ \text{Densidad} = \frac{2 \times \text{número de aristas}}{\text{número de nodos} \times (\text{número de nodos} - 1)} $$

En cambio, para redes dirigidas, la fórmula es ligeramente diferente porque las direcciones de las aristas permiten que cada par de nodos tenga hasta dos conexiones (una en cada dirección):
\[ \text{Densidad} = \frac{\text{número de aristas}}{\text{número de nodos} \times (\text{número de nodos} - 1)} \]

### Interpretación

- **Valor de Densidad 0**: Significa que no hay aristas entre los nodos, es decir, la red no tiene conexiones.
- **Valor de Densidad 1**: Indica que cada par de nodos está conectado por una arista, lo que resulta en una red completamente conectada.

Por ejemplo, en una red social, una alta densidad sugeriría que casi todos los miembros se conocen entre sí, mientras que una densidad baja indicaría que hay relativamente pocos vínculos entre los miembros comparado con el número total de posibles conexiones.

La densidad es útil para comparar la conectividad de redes que tienen un tamaño similar, pero debe interpretarse con cuidado cuando se comparan redes de diferentes tamaños, ya que redes más grandes tienden a tener una densidad más baja simplemente debido a que hay más pares de nodos entre los que podría existir una conexión.

# Grafos bipartitas

```python
from networkx.algorithms import bipartite

# Creamos un grafo bipartito
B = nx.Graph()
B.add_nodes_from(['U1', 'U2', 'U3', 'U4'], bipartite=0) # Agregamos los nodos del conjunto U
B.add_nodes_from(['V1', 'V2', 'V3'], bipartite=1) # Agregamos los nodos del conjunto V
B.add_edges_from([('U1', 'V1'), ('U1', 'V2'), ('U2', 'V2'), ('U3', 'V3'), ('U4', 'V2'), ('U4', 'V3')])

# Verificamos que el grafo es bipartito
assert bipartite.is_bipartite(B)

# Obtenemos los conjuntos de nodos U y V
U = set(n for n, d in B.nodes(data=True) if d['bipartite']==0)
V = set(B) - U

# Creamos las proyecciones U-V
U_projection = bipartite.projected_graph(B, U)
V_projection = bipartite.projected_graph(B, V)

print("Proyección U:")
print(U_projection.edges())

print("Proyección V:")
print(V_projection.edges())
```

```python
# Dibujamos el grafo
nx.draw(B, with_labels=True)
plt.show()
```

##Grafo multipartita

```python
# Crear un grafo multipartito (tripartito en este caso)
M = nx.Graph()
M.add_nodes_from(['U1', 'U2', 'U3'], bipartite=0) # Usuarios
M.add_nodes_from(['M1', 'M2', 'M3'], bipartite=1) # Películas
M.add_nodes_from(['G1', 'G2'], bipartite=2) # Géneros
# Agregar aristas entre usuarios y películas
M.add_edges_from([('U1', 'M1'), ('U2', 'M1'), ('U3', 'M2'), ('U1', 'M3'), ('U2', 'M3')])
# Agregar aristas entre películas y géneros
M.add_edges_from([('M1', 'G1'), ('M2', 'G1'), ('M3', 'G2')])

# Verificar si el grafo es bipartite (todos los grafos tripartitos son bipartitos)
assert bipartite.is_bipartite(M)
```

```python
# Dibujamos el grafo
nx.draw(M, with_labels=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Cargar el grafo de familias florentinas
G = nx.florentine_families_graph()

# Dibujar el grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # para un diseño consistente
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2500)

# Calcular y mostrar medidas de centralidad
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

''' # El hub en la red. Que nodo es el mas central.'''
print("Degree Centrality:") # El hub en la red. Que nodo es el mas central.
for k, v in degree_centrality.items():
    print(f"{k}: {v}")


''' Mide cuántas veces un nodo actúa como un "puente" a
lo largo del camino más corto entre dos otros nodos.
Los nodos con alta centralidad de intermediación pueden tener un control
significativo sobre la red porque la información debe pasar a través de ellos.'''

print("\nBetweenness Centrality:") #
for k, v in betweenness_centrality.items():
    print(f"{k}: {v}")


''' Mide la distancia promedio de un nodo a todos los demás nodos en la red.
Los nodos con alta centralidad de cercanía pueden difundir información
de manera eficiente a través de la red.'''

print("\nCloseness Centrality:")
for k, v in closeness_centrality.items():
    print(f"{k}: {v}")

'''Asigna puntuaciones relativas a todos los nodos en la red basándose en
el concepto de que las conexiones a nodos de alta puntuación contribuyen más
a la puntuación del nodo que las conexiones a nodos de baja puntuación.'''
print("\nEigenvector Centrality:")
for k, v in eigenvector_centrality.items():
    print(f"{k}: {v}")
```
